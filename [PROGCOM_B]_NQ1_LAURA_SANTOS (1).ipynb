**An치lisis experimental de la relaci칩n posici칩n-tiempo y velocidad-tiempo en un plano inclinado con fricci칩n y masa colgante.**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
time = np.linspace(0, 10, 100)
acceleration = 9.81 * np.sin(np.radians(30)) * 0.6                        # Se considera un factor de fricci칩n
position = 0.5 * acceleration * time ** 2 + np.random.normal(0, 0.5, 100) # Se agregan datos experimentales
velocity = acceleration * time + np.random.normal(0, 0.3, 100)            # Se agregan datos experimentales

plt.plot(time, position, 'o')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posici칩n (m)')
plt.title('Relaci칩n entre posici칩n y tiempo en un plano inclinado con fricci칩n y masa colgante (datos experimentales)')
plt.grid()
plt.show()

plt.plot(time, velocity, 'o')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Relaci칩n entre velocidad y tiempo en un plano inclinado con fricci칩n y masa colgante (datos experimentales)')
plt.grid()
plt.show()
```
